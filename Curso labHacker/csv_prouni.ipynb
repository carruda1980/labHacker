```python
# Importando bibliotecas
import pandas as pd
%matplotlib inline
```

```python
# Carregando dados
df = pd.read_csv('prouni.csv')
df.head()
```

```python
# Tipo dos dados
df.dtypes
```

```python
# Analisando os dados das colunas
df.describe()
```

```python
df.info()
```

```python
# Quantidade de cursos por uf
df['uf_busca'].value_counts()
```

```python
# Qual curso com mensalidade mais cara
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
# Qual o curso com mensalidade mais barato
df[df['mensalidade'] == df['mensalidade'].min()]
```

```python
# Qual mensalidade mais cara por uf
df.groupby('uf_busca')['mensalidade'].max()
```

```python
# Qual o curso com mensalidade mais barato por uf
df.loc[df.groupby(['uf_busca','nome'])['mensalidade'].idxmin()]
```

```python
# Quais os 5 cursos mais caros
df.nlargest(5, 'mensalidade')
```

```python
# Quais os 6 cursos mais baratos
df.nsmallest(6, 'mensalidade')
```
