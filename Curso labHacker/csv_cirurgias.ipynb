```python
# Importando as bibliotecas
import pandas as pd
# Informando ao matplotlib sobre o gráfico inline
%matplotlib inline
```

```python
# Para esse exemplo iremos utilizar apenas algumas colunas que são:
# 3 - Hospital, 6 - Município, 7 - Complexidade, 8 - Carater Atendimento
# 12 - Sub grupo atendimento, 14 - Procedimento

df = pd.read_csv('sih-janeiro-2017-cirurgias-eletiva-e-emergencia.csv', sep=';', encoding='cp1252', 
                 usecols=[3,6,7,8,12,14])
# Trocar nome das colunas
df.columns = ['Hospital', 'Município', 'Complexidade', 'Carater Atendimento', 'Sub grupo', 'Procedimento']
df.head()
```

```python
df.describe()
```

```python
# Listar os Hospitais presentes nos dados

df['Hospital'].unique()
```

```python
# Quantas cirurgias realizadas por hospital
df['Hospital'].value_counts()
```

```python
# Quantas cirurgias realizadas por sub grupo
df['Sub grupo'].value_counts()
```

```python
# Exibir as informações no gráfico
df['Sub grupo'].value_counts().plot.bar()
```

```python
# Calculando a porcentagem
total = df['Sub grupo'].count()
porcentagem = df['Sub grupo'].value_counts()/total
```

```python
porcentagem.plot.pie()
```

```python
# Realizar o filtro por um hospital espec'fico
df_hospital = df[df['Hospital'] == '0010529 HRPL']
df_hospital.head()
```

```python
# Quantidade de procedimento filtrando por alguma palavra
df_hospital[df_hospital['Procedimento'].str.contains('AMPUTA')].count()
```

```python
# Mudando a ordem dos dados no gráfico
# Adicionando titulo ao gráfico

df['Sub grupo'].value_counts(ascending = True).plot.bar(title='Gráfico por subgrupo')
```
